# DATA VISUALIZATION

```python
# Histogram of 2023 population
plt.figure(figsize=(10, 6))
sns.histplot(df_final["2023"], bins=30, kde=True, color='skyblue')
plt.title("Histogram of Country Populations (2023)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.grid(True)
plt.show()
```

```python
# Filter data for total population
total_population_data = df[df['indicator_code'] == 'SP.POP.TOTL']

# Sort data based on the total population for 2023
total_population_sorted = total_population_data.sort_values(by="2023", ascending=False)

# Get the top ten countries with the highest total population for 2023
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(total_top_ten_countries[['country_code']])
```

```python
# Bar Plot Top ten countries of total population in year 2023
plt.figure(figsize=(20, 10))
plt.subplot(2,2,1)
sns.barplot(x="2023", y="country_code", data=total_top_ten_countries, palette="viridis")
plt.title("Top Ten Countries of Total Population (2023)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
# Sort data based on the total population for 2023
total_population_sorted1 = total_population_data.sort_values(by="2023")

# Get the top ten countries with the highest total population for 2023
total_bottom_ten_countries = total_population_sorted1.head(10)
print("Bottom ten countries of total population\n")
print(total_bottom_ten_countries[['country_code']])
```

```python
# Bottom ten countries of total population in year 2023
plt.figure(figsize=(20, 10))
plt.subplot(2,2,1)
sns.barplot(x="2023", y="country_code", data=total_bottom_ten_countries, palette="cubehelix")
plt.title("Bottom Ten Countries of Total Population (2023)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
year_cols = df.columns[1:]
df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')
```

```python
country = 'India'

# Get row for the selected country, exclude the 'Country Name' column
country_data = df[df['country_name'] == country].iloc[:, 1:].T
country_data.columns = ['Population']
country_data.index.name = 'Year'

# Convert index to column and ensure it's string type
country_data.reset_index(inplace=True)
country_data['Year'] = country_data['Year'].astype(str)

# Plot
plt.figure(figsize=(12, 5))
sns.barplot(x='Year', y='Population', data=country_data, color='skyblue')
plt.xticks(rotation=90)
plt.title(f'{country} Population Over Time')
plt.show()
```

```python
country = 'India'
decades = ['1960', '1970', '1980', '1990', '2000', '2010', '2020', '2022']
pop_by_decade = df[df['country_name'] == country][decades].melt(var_name='Year', value_name='Population')

plt.figure(figsize=(10, 5))
sns.barplot(data=pop_by_decade, x='Year', y='Population', color='orange')
plt.title(f'{country} Population by Decade')
plt.show()
```

```python
countries = ['India', 'China', 'United States', 'Indonesia', 'Brazil']
compare_2022 = df[df['countryname'].isin(countries)][['Country Name', '2022']]

plt.figure(figsize=(8, 5))
sns.barplot(data=compare_2022, x='Country Name', y='2022', palette='Set2')
plt.title('Population of Selected Countries (2022)')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
